## Bloch Sphere

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to plot the Bloch sphere
def plot_bloch_sphere(save_path=None):
    # Create a sphere
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Define the range for the u and v variables
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    
    # Calculate the x, y, and z coordinates of the sphere
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))
    
    # Plot the surface of the sphere
    ax.plot_surface(x, y, z, color='b', alpha=0.1, rstride=5, cstride=5)
    
    # Plot the x, y, and z axes
    ax.quiver(0, 0, 0, 1, 0, 0, color='r', linewidth=1.5)
    ax.quiver(0, 0, 0, 0, 1, 0, color='g', linewidth=1.5)
    ax.quiver(0, 0, 0, 0, 0, 1, color='b', linewidth=1.5)
    
    # Plot labels for the axes
    ax.text(1.1, 0, 0, r'$|1\rangle$', color='r', fontsize=16)
    ax.text(-1.1, 0, 0, r'$|0\rangle$', color='r', fontsize=16)
    ax.text(0, 1.1, 0, r'$|+\rangle$', color='g', fontsize=16)
    ax.text(0, -1.1, 0, r'$|-\rangle$', color='g', fontsize=16)
    ax.text(0, 0, 1.1, r'$|i\rangle$', color='b', fontsize=16)
    ax.text(0, 0, -1.1, r'$|-i\rangle$', color='b', fontsize=16)
    
    # Set plot limits
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    
    # Set plot labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    # Save the plot if a save path is provided
    if save_path:
        plt.savefig(save_path)
    else:
        # Show the plot
        plt.show()

# Define the path to the desktop
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop', 'bloch_sphere.png')

# Call the function to plot the Bloch sphere and save it to the desktop
plot_bloch_sphere(desktop_path)
```

## Hermite Polynomials


```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite

# Function to calculate the Hermite wave function
def hermite_wave_function(n, x):
    Hn = hermite(n)
    normalization = 1.0 / np.sqrt(2**n * np.math.factorial(n) * np.pi**0.25)
    psi_n = normalization * Hn(x) * np.exp(-x**2 / 2)
    return psi_n

# Set up the x range and number of points
x = np.linspace(-4, 4, 1000)

# Plot the first few Hermite wave functions
plt.figure(figsize=(10, 6))

for n in range(5):  # Plotting first 5 wave functions
    psi_n = hermite_wave_function(n, x)
    plt.plot(x, psi_n, label=f'n={n}')

plt.title('Hermite Wave Functions for Quantum Harmonic Oscillator')
plt.xlabel('x')
plt.ylabel('$\psi_n(x)$')
plt.legend()
plt.grid(True)

# Define the path to save the plot on the desktop
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop', 'hermite_wave_functions.png')
plt.savefig(desktop_path)
plt.show()
```

## Wave function of Particle in box

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the wave function of a particle in a 1D box
def particle_in_box_wave_function(n, x, L):
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

# Set up the parameters
L = 1  # Length of the box
x = np.linspace(0, L, 1000)  # Position range

# Plot the first few wave functions
plt.figure(figsize=(10, 6))
for n in range(1, 5):  # Plotting first 4 wave functions
    psi_n = particle_in_box_wave_function(n, x, L)
    plt.plot(x, psi_n, label=f'n={n}')

plt.title('Wave Functions of Particle in a 1D Box')
plt.xlabel('x')
plt.ylabel('$\psi_n(x)$')
plt.legend()
plt.grid(True)

# Define the path to save the plot on the desktop
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop', 'particle_in_box_wave_functions.png')
plt.savefig(desktop_path)
plt.show()
```
